# Numerical simulation of coupled Lorenz's Equation on random graph by RK4.

```python
# import libraries.
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
# self-made module
from Coupled_Lorenz_chaos import Coupled_Lorenz_Sys
```

Now, numpber of Lorenz system is 100. It can be set by constructor of Coupled_Lorenz_Sys.

```python
N_LS = 100 
```

```python
# set the initial state
# initial stata can be set as numpy array ( ( index of system, xyz))

init_v = np.random.rand(N_LS, 3)
print(init_v.shape)
```

```python
# instantation
CLS = Coupled_Lorenz_Sys(5)
```

```python
# set init value
CLS.set_init_state(init_vec=init_v)
```

```python
CLS.set_const(C_cnst=0.1)
```

```python
print(CLS.C)
```

```python
# visualization of connection of system 
net_c = nx.from_numpy_matrix(CLS.C)
nx.draw_networkx(net_c, arrows=True, node_color="green")
#plt.show()
```

```python
# excute the numrical simulation.
XYZs, Ts = CLS.solve(0.01, 10000)
```

```python
fig, ax = plt.subplots(CLS.N_osc, 1, figsize=(15, 12))


for i in range(0, CLS.N_osc):
    for j in range(0, 3):
        ax[i].plot(Ts, XYZs[i, j, :])
```

```python
fig = plt.figure()
Ax = fig.add_subplot(111, projection='3d')
Ax.scatter(XYZs[1, 0,:], XYZs[1, 1,:], XYZs[1, 2,:], s=5,   c=Ts, cmap="winter")
# 
Ax.set_xlabel('x')
Ax.set_ylabel('y')
Ax.set_zlabel('z')
Ax.axis('equal')
```
